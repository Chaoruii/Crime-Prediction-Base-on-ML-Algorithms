```python
import pandas as pd
import numpy as np

xlabel=pd.read_csv("data/x.csv")
ylabel=pd.read_csv("data/y.csv")
PriceLabel=pd.read_csv("data/avg_record.csv")
CensusLabel=pd.read_csv("data/census_record.csv")
#AreaLabel=pd.read_csv("data/area.csv")
xlabel.head()

xlabel.drop(columns=['Unnamed: 0'],inplace = True)
ylabel.drop(columns=['Unnamed: 0'],inplace = True)
PriceLabel.drop(columns=['Unnamed: 0'],inplace = True)
CensusLabel.drop(columns=['Unnamed: 0'],inplace = True)
#AreaLabel.drop(columns=['Unnamed: 0'],inplace = True)

xv = np.array(xlabel)
yv = np.array(ylabel)
avg_record = np.array(PriceLabel)
census_record = np.array(CensusLabel)
#area_record = np.array(AreaLabel)


police =pd.read_csv("data/police_result.csv")
```

```python
avg_list =[]
census_list = []
#area_list = []

for i in police.index:
    idx = (np.abs(xv[0]-float(police["Longitude"][i]))).argmin()
    idy = (np.abs(yv[:,0]-float(police["Lattitude"][i]))).argmin()
    avg1 = avg_record[idy,idx]
    census = census_record[idy,idx]
#    area1 = area_record[idy,idx]
    avg_list.append(avg1)
    census_list.append(census)
#    area_list.append(area1)
```

```python
police ['average'] = avg_list
police ['census'] = census_list
#police ['area'] = area_list
```

```python
police.drop (police[police ['census'].isin([1])].index,inplace = True)
```

```python
PopuData = pd.read_csv('data/population.csv')
PopuData.drop(columns=['UAID'],inplace = True)
CensusNum = r'(\d+\.*\d*)'
Census_N=PopuData['Census'].str.extract(CensusNum, expand=False)
Census_replace=pd.DataFrame(Census_N).astype(float)
PopuData.drop(columns=['Census'],inplace = True)
PopuData = pd.merge(PopuData,Census_replace,how='left',left_index=True,right_index=True)
PopuData.rename(columns = {'Census':'census'}, inplace = True)
PopuData.info()
```

```python
result1 = pd.merge(police,PopuData,how = 'left',on = 'census')
```

```python
area1 = pd.read_csv('data/area.csv')
result = pd.merge(result1,area1,how = 'left',on = 'census')
```

```python
result = result.drop('0',axis = 1)
```

```python
# Get one hot encoding of columns B
one_hot = pd.get_dummies(result['Type'])
# Drop column B as it is now encoded
result2 = result.drop('Type',axis = 1)
result2 = result2.drop('0',axis = 1)
# Join the encoded df
result2 = result2.join(one_hot)
```
